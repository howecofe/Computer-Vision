# Image Processing


```
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```
def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```
img = cv2.imread('/content/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

display_img(img)
```

- Apply a binary threshold onto the image.

```
img = cv2.imread('/content/giraffes.jpg', 0)
```

```
ret, thresh = cv2.threshold(img, 128, 255, cv2.THRESH_BINARY)
```

```
display_img(thresh, 'gray')
```

- convert colorspace to HSV, and display it

```
img = cv2.imread('/content/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```
display_img(img)
```

- Create a low pass filter with a 4 by 4 kernel filled with values of 1/10(0.1), and then use 2D convolution to blur the image 

```
img = cv2.imread('/content/giraffes.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```
kernel = np.ones((4, 4), np.float32) / (10)
kernel
```

```
dst = cv2.filter2D(img, -1, kernel)
```

```
display_img(dst)
```

- Create a Horizontal Sobel Filter with a kernel size of 5 to the grayscale version of the image, and display the result

```
img = cv2.imread('/content/giraffes.jpg', 0)
img = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)

display_img(img, 'gray')
```

- Plot the color histograms for the RED, BLUE, and GREEN channel of the image.

```
img = cv2.imread('/content/giraffes.jpg')

b, g, r = img[:, :, 0], img[:, :, 1], img[:, :, 2]

b_hist = cv2.calcHist([img], [0], None, [256], [0, 256])
g_hist = cv2.calcHist([img], [1], None, [256], [0, 256])
r_hist = cv2.calcHist([img], [2], None, [256], [0, 256])
```

```
plt.figure()
plt.plot(b_hist, color='b')
plt.plot(g_hist, color='g')
plt.plot(r_hist, color='r')
plt.xlim([0, 256])
plt.title("Giraffes Histograms")

plt.show()
```
