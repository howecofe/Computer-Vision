# Video Processing

- TASK

  - Create a draw_circle function for the callback function

  - Use two events `cv2.EVENT_LBUTTONDOWN` and `cv2.EVENT_LBUTTONUP`
  
  - Use a boolean variable to keep track if the mouse has been clicked up and down based on the events above

  - Use a tuple to keep track of the x and y where the mouse was clicked.

  - You should be able to then draw a circle on the frame based on the x,y coordinates from the Event 

```
import numpy as np
import cv2

# Create a function based on a CV2 Event (Left button click)

# 마우스 콜백 함수
def draw_circle(event, x, y, flags, param):
    global center, clicked

    # 마우스클릭을 하면 center점을 추적함
    # 마우스가 released 됐는지 안됐는지 알기 위해 boolean 변수 사용
    if event == cv2.EVENT_LBUTTONDOWN:
        clicked = True
        center = x, y
        cv2.circle(param, center, 50, (255, 0, 0), -1)



center = (0, 0)
clicked = False

# Video 캡쳐
path_video = './opera_house.mp4'
cap = cv2.VideoCapture(path_video)


# 캡쳐된 프레임을 연결할 창 생성
cv2.namedWindow('frame')

# 마우스 클릭 함수를 draw_circle 함수와 연결
cv2.setMouseCallback('frame', draw_circle)

while True:

    # 비디오 캡쳐 읽기(read)
    ret, frame = cap.read()

    # 왼쪽 마우스버튼이 클릭됐는지 안됐는지 판별하는 조건문
    if clicked == True:
        x, y = center
        event = cv2.EVENT_LBUTTONDOWN
        # 클릭됐으면 원을 그려라
        draw_circle(event, x, y, None, frame)

    # 생성된 결과를 띄우기
    cv2.imshow('frame', frame)

    # q 키를 누르면 종료
    if cv2.waitKey(1) == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
