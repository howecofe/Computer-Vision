# Numpy and Image Assessment

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-white')

from google.colab.patches import cv2_imshow
```

```
img = cv2.imread('/content/dog.jpg')
```

```
print(img.shape)
```

```
plt.imshow(img)
plt.show()
```

```
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.show()
```

- Flip Upside Down and display it in the notebook

```
img_flip = cv2.flip(img_rgb, 0)

plt.imshow(img_flip)
plt.show()
```

- Draw an empty RED rectangle around dogs face and display it

```
img_rect = cv2.rectangle(img_rgb, (250, 90), (490, 270), (255, 0, 0), 5)

plt.imshow(img_rect)
plt.show()
```

- Draw a BLUE TRIANGLE in the middle of the image, and display it

```
img = cv2.imread('/content/dog.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

pts = np.array([[250, 270], [490, 270], [370, 90]], np.int32).reshape((-1, 2, 1))

img_tri = cv2.polylines(img_rgb, [pts], True, (0, 0, 255), 5)

plt.imshow(img_tri)
plt.show()
```

- fill in the triangle
  - `cv2.fillPoly()`

```
img_tri_fill = cv2.fillPoly(img_tri, [pts], (0, 0, 255))

plt.imshow(img_tri_fill)
plt.show()
```
